
# Task-1: Decision Tree Classification on Iris Dataset

## Objective:
To implement a Decision Tree classifier using the Iris dataset, visualize the decision path, and evaluate model performance.

## Dataset:
The Iris dataset contains:
- 150 samples
- 4 features (sepal & petal length/width)
- 3 classes (Setosa, Versicolor, Virginica)


```python

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris                              # Load the Iris dataset
from sklearn.model_selection import train_test_split                # Split the dataset into training and testing sets
from sklearn.tree import DecisionTreeClassifier, plot_tree          # Create a Decision Tree classifier model
from sklearn.metrics import classification_report, accuracy_score   # Calculate the accuracy of the model


# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names


# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


# Create a Decision Tree classifier model
model = DecisionTreeClassifier(random_state=42)
# Train the model on the training data
model.fit(X_train, y_train)


# Make predictions on the test data
y_pred = model.predict(X_test)
print("Classification Report:\n")
print(classification_report(y_test, y_pred, target_names=target_names))
# Calculate the accuracy of the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))


plt.figure(figsize=(16, 10))
# Visualize the decision tree
plot_tree(model,
          feature_names=feature_names,
          class_names=target_names,
          filled=True,
          rounded=True,
          fontsize=12)
plt.title("Decision Tree - Iris Dataset")
plt.show()

```


## Conclusion:
- Decision Tree model trained successfully.
- Visualization shows feature splits and decision rules.
- Model evaluated using accuracy and classification metrics.
